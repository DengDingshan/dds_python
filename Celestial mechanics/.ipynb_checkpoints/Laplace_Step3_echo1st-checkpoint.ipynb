```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import datetime
import time

import DDS_CMpack as DDS

%matplotlib inline
```

```python
echo1st_obs = pd.read_csv('echo1st_format.csv')
echo1st_obs
```

```python
echo1st_forecast = pd.read_csv('echo1st_forecast.csv')
echo1st_forecast
```

```python
# 残差、残差、残差

delta_AlphaCosDelta = echo1st_forecast['Alpha_d']*np.cos(echo1st_forecast['Delta']) \
                    - echo1st_obs['alpha']*np.cos(np.radians(echo1st_obs['delta']))
delta_Delta = echo1st_forecast['Delta_d'] - echo1st_obs['delta']

pd.DataFrame({'$ \Delta a\cos\delta $': np.array(delta_AlphaCosDelta),
             '$ \Delta\delta $ ':np.array(delta_Delta)})
```

```python
plt.figure()

plt.xlabel('Alpha')
plt.ylabel('Delta')

plt.scatter(echo1st_obs['alpha'],echo1st_obs['delta'],c = 'orange')
plt.scatter(echo1st_forecast['Alpha_d'],echo1st_forecast['Delta_d'],c = 'b')
# plt.scatter(delta_AlphaCosDelta,delta_Delta)
```

```python
plt.scatter(echo1st_forecast['Alpha'],echo1st_forecast['Delta'])
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = echo1st_forecast['x']
y = echo1st_forecast['y']
z = echo1st_forecast['z']

ax = plt.subplot(111, projection='3d')  # 创建一个三维的绘图工程
#  将数据点分成三部分画，在颜色上有区分度
ax.plot(x, y, z, c='y')  # 绘制数据点
ax.scatter(0,0,0,c= 'b',s = 10)

ax.set_zlabel('Z')  # 坐标轴
ax.set_ylabel('Y')
ax.set_xlabel('X')
plt.show()
```
